```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
from scipy.stats import binom
```

```python
#Number of atempts
n = 10

x = range(0,11)

#Number of exp
p = 0.5

#Array of possible outcomes
prob = binom.pmf(x, n, p)

#Set prop of plot
fig, binom_plot = plt.subplots(figsize = (10, 8))
binom_plot.set_xlabel("no of heads", fontsize = 16)
binom_plot.set_ylabel("Probability", fontsize = 16)
binom_plot.vlines(x, 0, prob, colors = 'r', lw = 5, alpha = 0.5)

#Plot
binom_plot.plot(x, prob, 'ro')
plt.show()
```

# CENTRAL LIMIT THEOREM (clt)
In the study of probability theory, the central limit theorem (CLT) states that the distribution of sample approximates a normal distribution (also known as a “bell curve”) as the sample size becomes larger, assuming that all samples are identical in size, and regardless of the population distribution shape.

```python
#clt mean of the random sample will closely resemble the mean of the whole population
#as the sample size increases regardless of shape of distribution
```

```python
#1000 random no
x = np.random.randint(0, 1000, size = (1, 1000))[0]
print(x.mean())

#30 dm samples from 25 dp
resamples = [np.random.choice(x, size = 25, replace = True) for i in range(30)]

avg_lst = []
for i in range(0,30):
    avg_lst.append(resamples[i].mean())

cumm_mean = sum(avg_lst)/len(avg_lst)
```
